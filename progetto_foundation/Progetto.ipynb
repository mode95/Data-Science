## 1. Convert the app sizes to a number and 
## 3. Transform “Varies with device” into a missing value

```python
app["Size"] = app["Size"].str.replace("M", "")
app["Size"] = app["Size"].str.replace(",", "")
```

```python
def app_size(string):
    if re.search("k$", string):
        return int(re.match("\d*", string).group()) / 1000
    else:
        re.search(".*", string)
        return re.match(".*", string).group()
```

```python
app["Size"] = app.apply(lambda row: app_size(row["Size"]), axis = 1)
```

```python
app = app.replace({"Varies with device" : np.nan})
app
```

```python
app["Size"] = pd.to_numeric(app["Size"])   # Ho trasformato Size in una variabile numerica con unità di misura
type(app["Size"][1])                       # Megabyte
```
